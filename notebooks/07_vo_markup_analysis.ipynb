```python
import pandas as pd
import matplotlib.pyplot as plt
caminho_do_arquivo = 'C:\\Users\\vinciaott\\OneDrive\\√Årea de Trabalho\\Tabela_com_selecao_features.csv'

df = pd.read_csv(caminho_do_arquivo)

print(df.columns)
```

```python
colunas_especificas = ['Area', 'Year', 'Cost of a healthy diet (PPP dollar per person per day)']
df_selecionado = df[colunas_especificas]
```

```python
df_selecionado.head()
```

```python
df_brazil = df_selecionado[df_selecionado['Area'] == 'Brazil']
df_brazil.head()
```

```python
# Convertendo a coluna 'Year' para inteiros e ordenando
df_brazil['Year'] = df_brazil['Year'].astype(int)
df_brazil = df_brazil.sort_values('Year')

# Plotando o gr√°fico
plt.figure(figsize=(10, 6))
plt.plot(df_brazil['Year'], df_brazil['Cost of a healthy diet (PPP dollar per person per day)'], marker='o', linestyle='-')

# Adicionando t√≠tulos e r√≥tulos
plt.title('Custo de uma Dieta Saud√°vel no Brazil ao Longo dos Anos', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Custo de uma Dieta Saud√°vel (PPP d√≥lar por pessoa por dia)', fontsize=12)
plt.grid(True)

# Ajustando os ticks do eixo x para mostrar todos os anos
plt.xticks(df_brazil['Year'], rotation=45)

# Exibindo o gr√°fico
plt.show()
```

```python
# Filtrando para o ano de 2021
df_2021 = df_selecionado[df_selecionado['Year'] == 2021]

# Encontrando os 10 maiores custos em 2021
df_maiores_custos_2021 = df_2021.nlargest(80, 'Cost of a healthy diet (PPP dollar per person per day)')

# Exibindo os 10 maiores custos em 2021
df_maiores_custos_2021
```

```python
# Lista de pa√≠ses da Am√©rica Latina
paises_latam = [
    'Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Costa Rica', 'Cuba', 'Dominican Republic',
    'Ecuador', 'El Salvador', 'Guatemala', 'Honduras', 'Mexico', 'Nicaragua', 'Panama', 'Paraguay', 
    'Peru', 'Uruguay', 'Venezuela'
]
# Selecionando as colunas espec√≠ficas
colunas_especificas = ['Area', 'Year', 'Cost of a healthy diet (PPP dollar per person per day)']
df_selecionado = df[colunas_especificas]

# Filtrando para a Am√©rica Latina e o ano de 2021
df_latam_2021 = df_selecionado[(df_selecionado['Year'] == 2021) & (df_selecionado['Area'].isin(paises_latam))]

# Encontrando os 10 maiores custos em 2021 na Am√©rica Latina
df_maiores_custos_latam_2021 = df_latam_2021.nlargest(13, 'Cost of a healthy diet (PPP dollar per person per day)')
```

```python
(df_maiores_custos_latam_2021)
```

```python
#Lista de pa√≠ses da Am√©rica Latina
paises_latam = [
    'Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Costa Rica', 'Cuba', 'Dominican Republic',
    'Ecuador', 'El Salvador', 'Guatemala', 'Honduras', 'Mexico', 'Nicaragua', 'Panama', 'Paraguay', 
    'Peru', 'Uruguay', 'Venezuela'
]


# Ordenando os dados para visualiza√ß√£o
df_maiores_custos_latam_2021 = df_maiores_custos_latam_2021.sort_values('Cost of a healthy diet (PPP dollar per person per day)', ascending=False)

# Criando uma lista de cores, destacando o Brasil
colors = ['yellow' if area == 'Brazil' else 'skyblue' for area in df_maiores_custos_latam_2021['Area']]

# Plotando o gr√°fico de barras
plt.figure(figsize=(12, 8))
plt.barh(df_maiores_custos_latam_2021['Area'], df_maiores_custos_latam_2021['Cost of a healthy diet (PPP dollar per person per day)'], color=colors)
plt.xlabel('Custo de uma Dieta Saud√°vel (PPP d√≥lar por pessoa por dia)', fontsize=12)
plt.ylabel('√Årea', fontsize=12)
plt.title('Os 13 Custos Mais Elevados de uma Dieta Saud√°vel na Am√©rica Latina em 2021', fontsize=14)
plt.gca().invert_yaxis()  # Inverter o eixo y para que o maior custo esteja no topo

# Adicionando os valores nas barras
for index, value in enumerate(df_maiores_custos_latam_2021['Cost of a healthy diet (PPP dollar per person per day)']):
    plt.text(value, index, f'{value:.2f}', va='center', ha='left', fontsize=10, color='black')

# Exibindo o gr√°fico
plt.show()
```
