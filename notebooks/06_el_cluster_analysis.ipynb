## 2.2 Data wrangling

### Preparação dos dados para a análise

Manter:
* Dados históricos de 2017 e 2021 dos países do cluster
* Dados não nulos

#### Colunas a serem consideradas

```python
df2 = df[(df['Area'].isin(df_cluster['Area'])) & (df['Year'].isin([2017, 2021]))].copy(deep=True)
```

```python
df_features = [ 'Area', 'Year',
'Cost of a healthy diet (PPP dollar per person per day)', 
'Coefficient of human inequality',
'Human Development Index (value)', 
'Inequality in eduation',
'Inequality in income', 
'Inequality in life expectancy',
'2.a.1 Agriculture share of Government Expenditure', 
'2.a.1 Agriculture value added share of GDP', 
'6.4.1 Water Use Efficiency (Agriculture (ISIC4 A01 A0210 A0322))', 
'6.4.1 Water Use Efficiency (Industries)',
'Gross domestic product per capita, PPP, (constant 2017 international $)',
'Per capita food supply variability (kcal/cap/day)', 
'Value of food imports in total merchandise exports (percent) (3-year average)',
'Prevalence of undernourishment (percent) (3-year average)',
'Percentage of the population unable to afford a healthy diet (percent)'] 

df_independents = [ 
'Cost of a healthy diet (PPP dollar per person per day)', 
'Coefficient of human inequality',
'Human Development Index (value)', 
'Inequality in eduation',
'Inequality in income', 
'Inequality in life expectancy',
'2.a.1 Agriculture share of Government Expenditure', 
'2.a.1 Agriculture value added share of GDP', 
'6.4.1 Water Use Efficiency (Agriculture (ISIC4 A01 A0210 A0322))', 
'6.4.1 Water Use Efficiency (Industries)',
'Gross domestic product per capita, PPP, (constant 2017 international $)',
'Per capita food supply variability (kcal/cap/day)', 
'Value of food imports in total merchandise exports (percent) (3-year average)',
'Prevalence of undernourishment (percent) (3-year average)',
'Percentage of the population unable to afford a healthy diet (percent)'] 

df3 = df2[df_features].dropna().copy(deep=True)
```

```python
df3
```

#### Ajuste dos tipos de dados

```python
df3.info()
```

```python
df3[['Area']] = df3[['Area']].astype('string')
df3[['Year']] = df3[['Year']].astype('int')
df3[df_independents] = df3[df_independents].astype('float')

df3.info()
```

## 2.3 Exploration Analysis

### Análise das Distribuições

```python
df3_2017 = df3[df3['Year'] == 2017].set_index('Area')

df3_2017 = df3_2017.sort_values(by='Percentage of the population unable to afford a healthy diet (percent)', ascending=False).reset_index()

analysis_col = 'Percentage of the population unable to afford a healthy diet (percent)'

plt.figure(figsize=(6,10))
plt.title('% da População incapaz de pagar uma dieta saudável (2017)')

plt.barh(df3_2017['Area'], df3_2017[analysis_col])

bar_list = plt.barh(df3_2017['Area'], df3_2017[analysis_col], color = ['lightgray'])

bar_list[15].set_color('green')

bar_list[4].set_color('blue')
bar_list[5].set_color('blue')
bar_list[11].set_color('blue')

plt.show()
```

```python
df3_2021 = df3[df3['Year'] == 2021].set_index('Area')

df3_2021 = df3_2021.sort_values(by='Percentage of the population unable to afford a healthy diet (percent)', ascending=False).reset_index()

analysis_col = 'Percentage of the population unable to afford a healthy diet (percent)'

plt.figure(figsize=(6,10))
plt.title('% da População incapaz de pagar uma dieta saudável (2021)')

plt.barh(df3_2021['Area'], df3_2021[analysis_col])

bar_list = plt.barh(df3_2021['Area'], df3_2021[analysis_col], color = ['lightgray'])

bar_list[7].set_color('green')

bar_list[15].set_color('blue')
bar_list[18].set_color('blue')
bar_list[19].set_color('blue')

plt.show()
```

```python
df3_2017 = df3[df3['Year'] == 2017].set_index('Area')
df3_2021 = df3[df3['Year'] == 2021].set_index('Area')

df3_delta = df3_2017.sub(df3_2021).dropna()
df3_delta.reset_index(inplace=True)
df3_delta = df3_delta.sort_values(by='Percentage of the population unable to afford a healthy diet (percent)')

df3_delta
```

```python
analysis_col = 'Percentage of the population unable to afford a healthy diet (percent)'

plt.figure(figsize=(15,8))
plt.title('Delta 2017-2021: % População incapaz de pagar uma dieta saudável')

bar_list = plt.barh(df3_delta['Area'], df3_delta[analysis_col], color = ['lightgray'])

bar_list[2].set_color('green')

bar_list[17].set_color('blue')
bar_list[18].set_color('blue')
bar_list[19].set_color('blue')

plt.show()
```

```python
analysis_col = 'Coefficient of human inequality' 

df3_delta = df3_delta.sort_values(by=analysis_col)

plt.figure(figsize=(15,8))
plt.title('Delta 2017-2021: Coeficiente de desigualdade humana')

bar_list = plt.barh(df3_delta['Area'], df3_delta[analysis_col], color = ['lightgray'])

bar_list[4].set_color('green')

bar_list[8].set_color('blue')
bar_list[18].set_color('blue')
bar_list[19].set_color('blue')

plt.show()
```

```python
analysis_col = 'Inequality in income'

df3_delta = df3_delta.sort_values(by=analysis_col)

plt.figure(figsize=(15,8))
plt.title('Delta 2017-2021: Desigualdade de renda')

bar_list = plt.barh(df3_delta['Area'], df3_delta[analysis_col], color = ['lightgray'])

bar_list[3].set_color('green')

bar_list[13].set_color('blue')
bar_list[18].set_color('blue')
bar_list[19].set_color('blue')

plt.show()
```

```python
analysis_col = 'Human Development Index (value)'

df3_delta = df3_delta.sort_values(by=analysis_col)

plt.figure(figsize=(15,8))
plt.title('Delta 2017-2021: IDH')

bar_list = plt.barh(df3_delta['Area'], df3_delta[analysis_col], color = ['lightgray'])

bar_list[8].set_color('green')

bar_list[5].set_color('blue')
bar_list[13].set_color('blue')
bar_list[17].set_color('blue')

plt.show()
```
